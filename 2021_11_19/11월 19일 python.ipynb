# list 내장 함수

## .extend() 

```python
list1 = [1,2,3,4,5]
list2 = [6,8,1234,346,7,345,]

#이렇게는 안되네 
list3 = list1.extend(list2)
print(list3)

#list1에다가 list2를 붙여넣는 것 
list1.extend(list2)
print(list1)
```

## .pop

```python
list = [1,23,2,4,56,23,4235,2345,3456,345,]
ouput = list.pop(0) #pop을 이용해서 리스트안의 값을 빼올 수 있다
print(ouput) 
print(list) # 빼온 요소를 제외한 리스트가 출력된다.
```

## .reverse() 

```python
list = [1,3,34,1,23,45,3,6,7,2]
list.reverse() #끝에 있는 인덱스를 앞으로 해서 가지고 역순정렬하는 구나
print(list)
```

## .sort()

```python
list = [11,1,3,5,1,5,2,36,345]
list.sort(reverse=True) #sort에서 reverse는 요소의 크기를 큰것부터 역순정렬
print(list)
```

```python
list = [11,1,3,5,1,5,2,36,345]
list.sort() #요소의 크기 순으로 정렬
print(list)
```

dic.itmes()

```python
phone_book = {"홍길동":"010-1233-1231",
             "강감찬":"010-1512-1512",
             "이순신":"010-1231-1234"}

for k,v in phone_book.items():
    print("{}의 전화번호는 {} 입니다".format(k,v))
```

### 강수량 그래프 만들기

```python
from matplotlib import pyplot as plt

month = [1,2,3,4,5,6]
precipitation = [24.6, 24.9, 97.5, 50.4, 74.9, 105.2]

plt.plot(month, precipitation,marker=',')
plt.title("2009년 대관령 월평균 강수량")
plt.xlabel("월")
plt.ylabel("강수량(mm)")
#plt.legend([]) #그래프 요약정보를 작게 띄워주는거 여러개 가능하니까 리스트로 받네
plt.show()
```

```python
from matplotlib import font_manager,rc    
    
font_list=font_manager.findSystemFonts() 

    
# 이름순정렬
font_list.sort() 

for file_path in font_list:

        #폰트 파일의 경로를 사용하여 폰트 속성 객체 가져오기
        fp=font_manager.FontProperties(fname=file_path) 


        # 폰트 속성을 통해 파이썬에 설정해야 하는 폰트 이름 조회 
        font_name=fp.get_name() 


        # 폰트 파일 경로와 폰트 이름 출력하기
        print("%s >> %s" %(file_path,font_name))
```
