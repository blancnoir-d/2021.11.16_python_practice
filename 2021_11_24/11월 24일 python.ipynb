```python
import numpy as np
#본래 선형(1차원)을 되어 있는걸 차원을 달리해서 인지할 수 있도록 하는데 numpy
array = np.arange(8).reshape(2,4)
print(array)

#left, right = np.split(array, [3], axis=0) #해당 array를 인덱스 3을 기준으로 가로로(행)으로 나뉨
left, right = np.split(array, [2], axis=1) #해당 array를 인덱스 2(2앞에서 잘림)를 기준으로 세로로(열)로 나뉨

print("left",left)
print("모양",left.shape)
print("right",right)
print(right.shape)
```

```python
#시리즈를 이용해서 데이터 프레임 만들기 

#딕셔너리
word_dict = {
'Apple': '사과',
'Banana': '바나나',
'Carrot': '당근',
'Durian': '두리안'
}
frequency_dict = {
'Apple': 3,
'Banana': 5,
'Carrot': 7,
'Durian': 2
}
importance_dict = {
'Apple': 3,
'Banana': 2,
'Carrot': 1,
'Durian': 1
}

#딕셔너리로 시리즈로 만듦 
word = pd.Series(word_dict)
frequency = pd.Series(frequency_dict)
importance = pd.Series(importance_dict)

Df = pd.DataFrame({'word':word, 'frequency':frequency, 'importance':importance})
print(Df)
```

```python
listTest = [word, frequency, importance]
print(listTest)
```

```python
# 연산 & 새로운 컬럼 생성 
score = Df['frequency'] * Df['importance']
Df['score'] = score
print(Df)
```

```python
#데이터프레임 슬라이싱
print(Df.loc['Banana':'Carrot','importance':]) #바나나에서 캐럿까지 importance포함 이후까지 
```

```python
#행 조회하기
#이 상태로 하면 1차원 시리즈 형태가 리턴 
print(Df.loc['Apple'])
```

```python
#배열 형태를 입력하면 2차원 데이터 프레임 형태로 출력
print(Df.loc[['Apple']])
```

```python
#여러행 조회
print(Df.loc[['Apple','Banana']])
```

```python
#인덱스를 기준으로 슬라이싱 
print(Df.iloc[1:3, 2:])
```

```python
#데이터 프레임의 마스킹
import pandas as pd
import numpy as np


#이런식으로 numpy를 이용해서 데이터 프레임을 만드는 거에 익숙해져야 할 듯 
df = pd.DataFrame(np.random.randint(1,10,(2,2)), index=[0,1], columns=["A","B"])

#데이터 프레임 출력하기
print(df)
print()

print(df['A']<=5) # 컬럼A의 각 원소가 5보다 작거나 같은지 출력 

#컬럼 A의 원소가 5보다 작고, 컬럼 B의 원소가 8보다 작은 행 추출 
print()
print(df.query("A<=5 and B <=8"))
```

```python
#데이터 프레임의 개별 연산1 

df = pd.DataFrame([[1, 2, 3, 4], [1, 2, 3, 4]], index=[0, 1], columns=["A", "B", "C", "D"])
print(df)
print()
df = df.apply(lambda x: x + 1)
print(df)
print()



def addOne(x):
    return x + 1


df = df.apply(addOne)
print(df)
```

```python
#replace

df = pd.DataFrame([
['Apple', 'Apple', 'Carrot', 'Banana'],
['Durian', 'Banana', 'Apple', 'Carrot']],
index=[0, 1],
columns=["A", "B", "C", "D"])
print(df)
df = df.replace({"Apple": "Airport"})
print(df)
```

```python
#groupby

df = pd.DataFrame([
['Apple', 7, 'Fruit'],
['Banana', 3, 'Fruit'],
['Beef', 5, 'Meal'],
['Kimchi', 4, 'Meal']],
columns=["Name", "Frequency", "Type"])
print(df)
print(df.groupby(['Type']).sum())
```
