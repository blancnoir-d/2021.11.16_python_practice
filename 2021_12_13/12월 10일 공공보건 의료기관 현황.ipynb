### 표준 이름으로 변경 

```python
addr_aliases = {'경기':'경기도', '경남':'경상남도', '경북':'경상북도', '충북':'충청북도', '서울시':'서울특별시', '부산특별시':'부산광역시', '대전시':'대전광역시', '충남':'충청남도', '전남':'전라남도', '전북':'전라북도'}

#이름 변경
addr['시도'] = addr['시도'].apply(lambda v: addr_aliases.get(v,v))
```
